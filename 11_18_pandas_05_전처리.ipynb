```python
# concat 과 merge
# concat : row나 column 기준으로 단순하게 이어 붙히기
# merge : 특정 고유한 키값을 기준으로 병합
df2
```

```python
# pd.merge(left, right, on='기준column', how='left')
# left와 right는 병합할 두개의 dataFrame
# on에는 병합의 기준이 되는 column
# how에는 left, right, inner, outer 4개의 병합방식중 선택
df_right = df2.drop([1,3,5,6,9], axis=0)
df_right = df_right.reset_index(drop=True)
df_right
```

```python
pd.concat([df, df_right], axis=1)
```

```python
pd.merge(df, df_right, on='이름', how='left')
```

```python
pd.merge(df, df_right, on='이름', how='right')
```

```python
pd.merge(df, df_right, on='이름', how='inner')
```

```python
pd.merge(df, df_right, on='이름', how='outer')
```

```python
#df와 df_right를 병합, 성명으로 표기되어 있는 column은 df의 이름과 같은 성격의 데이터
pd.merge(df,df_right,left_on='이름',right_on='성명',how='outer')
```
