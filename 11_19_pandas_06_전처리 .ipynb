```python
# apply에 활용할 수 있는 함수를 정의
def male_or_female(x):
  if x == '남자':
    return 1
  elif x == '여자':
    return 0
```

```python
df['성별_new'] = df['성별'].apply(male_or_female) # 시리즈에 apply를 적용시 axis는 필요 없음
```

```python
# cm당 브랜드평판지수를 구하기(브랜드평판지수 / 키)
def cm_to_brand(df):
  value = df['브랜드평판지수'] / df['키']
  return value
```

```python
df.apply(cm_to_brand)
```

```python
df.apply(cm_to_brand, axis=1) # 데이터프레임 전체에 대한 apply를 지정시 axis를 지정해준다.
```
