```python
# DataFrame
# 리스트를 이용해서 DataFrame 만들기
company1 = [['삼성', 2000, '스마트폰'],
            ['현대', 1000, '자동차'],
            ['네이버', 500, '포털']]
```

```python
pd.DataFrame(company1)
```

```python
df1 = pd.DataFrame(company1)
```

```python
df1
```

```python
df1.columns = ['기업명', '매출액', '업종']
```

```python
df1
```

```python
# dict 이용하여 DataFrame 만들기
company2 = {'기업명' : ['삼성', '현대', '네이버'],
            '매출액' : [2000, 1000, 500],
            '업종' : ['스마트폰', '자동차', '포털']}
```

```python
df2 = pd.DataFrame(company2)
```

```python
df2
```
