```python
# 원핫인코딩(One-hot-encoding)
# 한개의 요소는 True 그리고 나머지 요소는 False로 만들어주는 기법
df.head()
```

```python
# 머신러닝을 돌리기 위해서 수치형으로 바꿔주기
blood_map = {
    'A': 0,
    'B': 1,
    'AB': 2,
    'O':3
}
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

```python
# 혈액형 코드를 머신러닝 알고리즘에 그대로 넣어 예측하라고 지시한다면 컴퓨터는 각각의 값들의
# 관계를 스스로 형성하게된다.
# 이 때 B형과 AB형을 더하면 O형이라고 하는 잘못된 관계를 맺을 수 있다.
# 이런 상황을 방지하기 위해서 각 값들이 독립적이라는 것을 알려줘야 한다..
# 이때 사용하는 방법이 원핫인코딩이다.

df['혈액형_code']
```

```python
# 원핫인코딩으로 만들어주는 함수(get_dummies())
pd.get_dummies(df['혈액형_code'])
```

```python
pd.get_dummies(df['혈액형_code'], prefix='혈액형')
```
